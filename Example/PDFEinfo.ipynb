## Data cleaned for graphing

```python
## df_indicator: dropped ID, AGE, Height, Gender, Handedness, More affected Side, L dopa equivalents, sessions2,3
## session 1 kept

df_indicators = df.drop(df.columns[np.r_[0:5, 6:11, 17: 20, 21, 23:25, 28:71]], axis=1)
df_indicators
```

```python
# Adding complete session 2 / 3 as new patient on the bottom
new_patients = []
drop_indices = [0, 7, 8, 9, 11, 13, 14]

def contains_empty(row, placeholder="-"):
    return any(item == placeholder for item in row)

for patient in range(len(df)):
    disease_duration = df.iloc[patient, 5]
    s2 = df.iloc[patient, 28:46].drop(df.columns[28:46][drop_indices])
    s3 = df.iloc[patient, 46:64].drop(df.columns[46:64][drop_indices])

    if  not contains_empty(s2):
        new_row = [disease_duration] + s2.tolist()
        new_patients.append(new_row)
        
    if not contains_empty(s3):
        new_row = [disease_duration] + s3.tolist()
        new_patients.append(new_row)

columns = ['Disease duration (years)'] + df.columns[10:28].drop(df.columns[10:28][drop_indices]).tolist()

new_patients_df = pd.DataFrame(new_patients, columns=columns)
new_patients_df = new_patients_df.apply(pd.to_numeric, errors='coerce')

df_indicators = pd.concat([df_indicators, new_patients_df], ignore_index=True).drop_duplicates().reset_index(drop=True)

pd.set_option('display.max_rows', None)
df_indicators
```

SVM

```python
from sklearn import svm
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
X = df_indicators[[col for col in df_indicators]]
# y = le.fit_transform(df_indicators['ID']).astype(float)
y = [ i for i in range(0,71)]

```

```python
PatientsModel = svm.SVC()
PatientsModel.fit(X,y)
```

K-means clustering

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
X = df_indicators[['Session 1 - UPDRS-II (score)', 'Session 1 - UPDRS-III (score)']]

# Graph the X values compared to the Y values
plt.scatter(X['Session 1 - UPDRS-II (score)'],X['Session 1 - UPDRS-III (score)'], c = y, cmap = 'gist_rainbow')
plt.xlabel("UPDRS-II", fontsize = 18)
plt.ylabel("UPDRS-III", fontsize = 18)
```

```python
km = KMeans(n_clusters = 35, random_state = 0)

km.fit(X)
```

```python
new_labels = km.labels_
plt.title("Predicted", fontsize = 18)
plt.scatter(X['Session 1 - UPDRS-II (score)'],X['Session 1 - UPDRS-III (score)'], c = new_labels, cmap = 'gist_rainbow')
plt.xlabel("Session 1 - UPDRS-II", fontsize = 18)
plt.ylabel("Session 1 - UPDRS-III", fontsize = 18)
```
