```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv('Results/stats-chosen/all_stats_angularAcceleration_fogRatio_updrsIII.csv')
```

```python
axis = df[['mean','std', 'FoG Ratio']]

plt.scatter(axis['std'], axis['FoG Ratio'])
plt.xlabel("Angular Acceleration Stdv", fontsize = 18)
plt.ylabel("FoG Ratio", fontsize = 18)
```

```python
plt.scatter(axis['mean'], axis['FoG Ratio'])
plt.xlabel("Angular Acceleration Mean", fontsize = 18)
plt.ylabel("FoG Ratio", fontsize = 18)
```
